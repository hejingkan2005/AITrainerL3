```python
import pandas as pd

data = pd.read_excel("智能音箱数据集.xlsx")

print(data.head())
row_count = data.shape[0]
user_count = data['用户ID'].nunique()
print(f"总行数: {row_count}")
print(f"总用户数: {user_count}")

# 用户使用习惯：分析哪些功能最常被使用；
# 时段分析
data['时间戳'] = pd.to_datetime(data['时间戳'])
data['小时'] = data['时间戳'].dt.hour
hourly_usage = data['小时'].value_counts().sort_values(ascending=False)
print(hourly_usage)


# 功能使用频率：识别最受欢迎的功能和较少使用的功能；
freq = data['功能调用类型'].value_counts().sort_values(ascending=False)
# add percentile
freq_percentile = freq / freq.sum() * 100
freq = pd.DataFrame({'频率': freq, '百分比': freq_percentile})
print(freq)

# 响应时间：考察不同功能的平均响应时间，找出可能的瓶颈
response = data.groupby('功能调用类型')['响应时间'].mean().sort_values(ascending=False)
print(response)
```
