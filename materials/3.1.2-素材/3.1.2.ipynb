```python
import pandas as pd

data = pd.read_excel('智能照明系统数据集.xlsx')

data['数据记录的时间戳'] = pd.to_datetime(data['数据记录的时间戳'], errors='coerce')
data['hour'] = data['数据记录的时间戳'].dt.hour
hourly_usage = data.groupby('hour')['光线亮度值（0-100）', '色温值（1000K-6500K）'].mean().sort_index(ascending=False)
print(hourly_usage)

```
