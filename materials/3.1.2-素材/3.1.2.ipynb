```python
import pandas as pd

data = pd.read_excel('智能照明系统数据集.xlsx')

# # 用户使用习惯：分析用户在一天中不同时段对灯光亮度的偏好；
# data['hour'] = data['数据记录的时间戳'].dt.hour
# hourly_brightness_preferences = data.groupby('hour')['光线亮度值（0-100）'].mean().sort_index()
# print(hourly_brightness_preferences)

# # 用户使用习惯：分析用户在一天中不同时段对色温值的偏好；
# color_category = ['深黄 / 红', '暖黄光', '暖白 / 中性白', '正白光', '冷白光']
# color_value_category = [1000, 2000, 3000, 4000, 5500, 6500]
# data['color'] = pd.cut(data['色温值（1000K-6500K）'], bins=color_value_category, labels=color_category)
# data['hour'] = data['数据记录的时间戳'].dt.hour
# hourly_color_preferences = data.groupby('hour')['color'].value_counts().unstack()
# print(hourly_color_preferences)

# 用户使用习惯： 分析用户在一天中不同时段对灯光亮度和色温值的偏好；
# split one day hours to 3 time periods: morning (6-12), afternoon (12-18), evening (18-24)
print(f"\n\n用户使用习惯:")
data['hour'] = data['数据记录的时间戳'].dt.hour
data['time_period'] = pd.cut(data['hour'], bins=[6, 12, 18, 24], labels=['早晨', '下午', '晚上'], right=False)
# 计算每个时间段的平均亮度值和平均色温值
brightness_color_preferences_by_time_period = data.groupby('time_period')[['光线亮度值（0-100）', '色温值（1000K-6500K）']].mean().round(1)
print(brightness_color_preferences_by_time_period)


# 功能使用频率: 识别哪些智能场景被频繁使用，哪些较少使用；
print(f"\n\n功能使用频率:")
feature_usage_frequency = data['使用的场景'].value_counts()
feature_usage_frequency_with_percentile = pd.DataFrame({'场景次数': feature_usage_frequency, '占比(%)': feature_usage_frequency / feature_usage_frequency.sum() * 100})
print(feature_usage_frequency_with_percentile)

# # 响应时间：考察智能照明系统在接收到用户指令后至灯光调整完成之间的平均响应时间，找出可能的延迟瓶颈
# avg_response_by_userId = data.groupby('用户标识符')['响应时间（秒）'].mean().sort_values(ascending=False)
# print(avg_response_by_userId)

# 响应时间：考察智能照明系统在接收到用户指令后至灯光调整完成之间的平均响应时间，找出可能的延迟瓶颈
# display avg response time, the shortest response time, the longest response time, and the median response time, and the standard deviation of response time
print(f"\n\n总体相应时间统计:")
print(f"平均响应时间：", data['响应时间（秒）'].mean().round(2))
print(f"最短响应时间：", data['响应时间（秒）'].min().round(2))
print(f"最长响应时间：", data['响应时间（秒）'].max().round(2))
print(f"中位数响应时间：", data['响应时间（秒）'].median().round(2))
print(f"响应时间标准差：", data['响应时间（秒）'].std().round(2))

response_time_by_scenarios = data.groupby('使用的场景')['响应时间（秒）'].mean().round(2).sort_values(ascending=False).to_frame('平均响应时间（秒）')
response_time_by_scenarios.index.name = '场景模式'
print(response_time_by_scenarios)
```
