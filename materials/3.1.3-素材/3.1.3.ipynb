```python
import pandas as pd

step_stat = pd.read_excel('智能健康手环数据集.xlsx', sheet_name='步数统计记录')
func_check_stat = pd.read_excel('智能健康手环数据集.xlsx', sheet_name='功能查看记录')

# 用户活动模式：分析用户在一周内不同时间段的活动水平，识别高峰时段和低谷时段。
step_stat['hour'] = pd.to_datetime(step_stat['日期'], errors='coerce').dt.hour
hourly_step = step_stat.groupby('hour')['步数'].mean()
step_weekly = pd.to_datetime(step_stat['日期'], errors='coerce').dt.day_name()
daily_step = step_stat.groupby(step_weekly)['步数'].mean()

print(hourly_step)
print(daily_step)

# 健康指标关注度：识别哪些健康指标（如步数、心率、睡眠时长）最受用户关注，哪些较少被查看
func_check_freq = func_check_stat['功能调用类型'].value_counts()
print(func_check_freq)

# 数据同步性能：评估手环与手机应用之间数据传输的平均延迟，找出影响同步传输速度的因素
avg_response_time_by_func_check = func_check_stat.groupby('功能调用类型')['传输延迟时间'].mean().round(1)
print(avg_response_time_by_func_check)

# 
```
